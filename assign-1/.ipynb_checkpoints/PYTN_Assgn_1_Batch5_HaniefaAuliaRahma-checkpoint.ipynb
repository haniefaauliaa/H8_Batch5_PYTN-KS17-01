# Data Exploration

## Rumusan pertanyaan yang ingin diketahui

2. Kejahatan apa saja yang sering terjadi?

```python
#menampilkan baris awal data (atas)
df.head()
```

```python
#mengetahui dtype per kolom
df.info()
```

```python
#melihat ringkasan dari setiap kolom
df.describe()
```

```python
#melihat missing value per kolom dan menjumlahkannya
df.isnull().sum()
```

```python
#mem filter list tahun 'Year'
df.year
```

```python
#mem-filter kolom 'year' hanya 2011 hingga 2016
filtered_year = df[(df['year'] >= 2011) & (df['year'] <= 2016)]

filtered_year
```

```python
#mem-filter kolom yang memiliki nilai 'value' lebih dari 0
years_with_values = filtered_year.loc[filtered_year['value'] > 0]

years_with_values
```

```python
#mengetahui daftar unique per major category berdasarkan 'filtered_year'
major_unique = years_with_values['major_category'].unique()

major_unique
```

```python
#mengelompokkan data berdasarkan 'major_category' dan menghitung jumlah nilai 'value' serta menampilkan value > 0
major_values = years_with_values.groupby('major_category')['value'].sum()

major_values
```

```python
#mengetahui daftar unique per major category berdasarkan 'filtered_year'
minor_unique = years_with_values['minor_category'].unique()

minor_unique
```

```python
#mengelompokkan data berdasarkan 'minor_category' dan menghitung jumlah nilai 'value' serta menampilkan value > 0
minor_values = filtered_year[filtered_year['value'] > 0].groupby('minor_category')['value'].sum()


minor_values
```

```python
#filter hanya baris dengan nilai "value" lebih dari 0
filtered_data = df[df['value'] > 0]

#menggabungkan kolom 'major_category' dan 'minor_category' menjadi satu teks
filtered_data['combined_categories'] = filtered_data['major_category'] + ' ' + filtered_data['minor_category']

#menggabungkan teks untuk 200 baris pertama
combined = ' '.join(filtered_data['combined_categories'][:500])

combined
```

```python
#mengelompokkan nilai 'value' kriminal dengan tahun 'year' dan bulan 'month'
monthly_crime = filtered_year.groupby(['year', 'month'])['value'].sum().reset_index()

monthly_crime
```

```python
#mengetahui kriminal tersering terjadi pada tahun 'year' dan bulan 'month' apa
most_common = monthly_crime[monthly_crime['value'] == monthly_crime['value'].max()]

most_common
```

```python
#menghitung jumlah kriminal per wilayah 'borough'
crime_per_borough = years_with_values.groupby('borough')['value'].sum().sort_values(ascending=False)

crime_per_borough
```

```python
#menghitung jumlah kriminal per wilayah 'borough'
crime_per_borough = years_with_values.groupby('borough')['value'].sum().sort_values(ascending=False)

crime_per_borough
```

```python
#mengelompokkan data berdasarkan 'borough', 'major_category', dan 'minor_category', dan menghitung jumlah kriminal yang terjadi
crime_by_category = years_with_values.groupby(['borough', 'major_category', 'minor_category'])['value'].sum().reset_index()

crime_by_category
```

```python
#menampilkan 5 wilayah dengan tingkat kriminal tertinggi
top5_high_crime = crime_per_borough.head(5)

top5_high_crime
```

```python
#melihat kriminal apa saja yang terjadi di salah satu top 5 wilayah dengan tingkat kriminal tertinggi yaitu Westminster
westminster = years_with_values[years_with_values['borough'] == 'Westminster']
type_of_crime = westminster['major_category'].unique()

type_of_crime
```

```python
type_of_crime_detail = westminster['minor_category'].unique()

type_of_crime_detail
```

```python
#melihat jumlah kejadian dari jenis - jenis kriminal yang ada di Westminster
total_type_of_crime = westminster.groupby('major_category')['value'].sum()

total_type_of_crime
```

```python
top5_boroughs = top5_high_crime.index

for borough in top5_boroughs:
    #filter data untuk wilayah tertentu
    borough_data = years_with_values[years_with_values['borough'] == borough]
    
    #ambil jenis kriminal berdasarkan kategori utama 'major_category'
    type_of_crime = borough_data['major_category'].unique()
    
    #menampilkan wilayah dan jenis kriminal
    print(f"Wilayah: {borough}")
    print(f"Jenis Kriminal: {', '.join(type_of_crime)}")
    print("\n")
```

# Data Visualization

## Area Plot

```python
#melihat perubahan 'value' atau jumlah kriminal seiring berjalannya waktu
plt.figure(figsize=(10, 6))
sns.lineplot(data=years_with_values, x='year', y='value', hue='major_category', palette='Set3')
plt.title('Perubahan Jumlah Kriminal Berdasarkan Major Category (2011-2016)')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Kriminal')

#mengatur posisi legend di luar plot
plt.legend(title='Major Category', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

#### summary :
Visualisasi menggunakan line plot diatas menunjukkan perubahan jumlah kriminal dari tahun 2011 hingga 2016. Dapat kita lihat bahwa terjadi kenaikan atau penurunan dalam jumlah kriminal (value) dalam berbagai kategori utama (major_category). Tentunya, terdapat kenaikan yang terlihat **signifikan pada tahun 2012** dengan perbuatan kriminal 'theft and handling' atau **pencurian**

```python
#melihat distribusi Kriminal berdasarkan Major Category dan Minor Categpry

#memilih beberapa kategori untuk visualisasi
filtered_category = ['Burglary', 'Violence Against the Person', 'Theft and Handling', 'Criminal Damage']
subset_data = crime_by_category[crime_by_category['major_category'].isin(filtered_category)]

plt.figure(figsize=(12, 8))
sns.set(style="whitegrid")

sns.barplot(x='value', y='minor_category', hue='major_category', data=subset_data, palette='Set3')
plt.title('Distribusi Kriminal berdasarkan Major Category dan Minor Category', pad=20)
plt.xlabel('Number of Crimes')
plt.ylabel('Minor Category of Crime')

plt.xticks(rotation=45, ha='right')
plt.legend(title='Major Category', loc='upper right')

plt.tight_layout()

plt.show()
```

#### summary :
Visualisasi menggunakan diagram batang grup (grouped bar plot) memberitahukan kita besaran kriminal berdasarkan kategori utama (major_category) dan kategori khusus (minor_kategory). Sehingga kita dapat mengetahui jenis kriminal yang mendominasi dan termasuk ke dalam kategori utama yaitu **'theft and handling'** seperti pencurian kendaraan, pencurian di toko dsb.

### Pie Chart

```python
#mengetahui berapa persen kontribusi dari tingkat kriminal per 'major_category'

#data jumlah kriminal per 'major_category'
major_category_values = years_with_values['major_category'].value_counts()

plt.figure(figsize=(7, 7))
plt.pie(major_category_values,
        labels=major_category_values.index,
        autopct='%1.1f%%',
        startangle=140,
        colors=plt.cm.Set3(range(len(major_category_values))))
plt.title('Distribusi Tingkat Kriminal per Major Category', pad=25)

#mengatur posisi legend di luar plot
plt.legend(major_category_values.index, title='Major Category', bbox_to_anchor=(1, 0.5), loc='center left')
plt.axis('equal') #membuat lingkaran sempurna

plt.show()
```

#### summary :
Visualisasi menggunakan pie chart untuk mengetahui kontribusi tingkat kejahatan per Kategori Utama 'major_category' dengan  persentase yang tersedia, membantu kita memahami sejauh mana masing-masing kategori berperan dalam total jumlah kejahatan.
